```python
# READ THE CSV FILE + REMOVE CUSTOMER ID COLUMN

loans_data = pd.read_csv('loans.csv')
#loans_data

loans_df = pd.DataFrame(loans_data) 
#loans_df

loans_df = loans_df.drop('Customer ID', axis=1) #drop the customerid from columns
loans_df
```

```python
# PRINT THE HEAD OF THE DATA
loans_df.head()
```

```python
# REMOVE ROWS FROM DATA THAT HAVE A TOO LARGE OF A LOAN

loans_df = loans_df[loans_df['Current Loan Amount'] < 99999999] # bool check if the Current Loan Amount is smaller than 99999999 and if not then removes the row
loans_df
```

```python
# Remove rows that have the annual income as NaN (not a number)
loans_df = loans_df.dropna(subset=['Annual Income']) # subset specifies that this applies only to Annual Income column
loans_df.head()
```

```python
# Get the average Current Loan Amount
loans_df_mean = loans_df['Current Loan Amount'].mean() # self explanatory :D
loans_df_mean
```

```python
# Get the highest and lowest Annual Income in the dataset
loans_df_highest_AI = loans_df['Annual Income'].max()
loans_df_lowest_AI = loans_df['Annual Income'].min()

print(f"highest: {loans_df_highest_AI}, min: {loans_df_lowest_AI}")
```

```python
# Get the Home Ownership value of the Loan ID = bbf87a87-22cd-4d10-bd9b-7a9cc1b6e59d
loans_df_loan_id = loans_df[loans_df['Loan ID'] == 'bbf87a87-22cd-4d10-bd9b-7a9cc1b6e59d'] # first take the wanted loan id row to new variable

loans_df_home_ownership_by_loan_id = loans_df_loan_id['Home Ownership'].iloc[0] #iloc[0] selects the first index assuming there is only 1 when loan id == unique LATER FOUND OUT LOAN ID IS NOT UNIQUE, not gonna change this took too much time to find the way :D
loans_df_home_ownership_by_loan_id
```

```python
# Create a new field into your dataset called Actual Annual Income. (formula: Annual Income – 12 * Monthly Debt)
loans_df['Actual Annual Income'] = loans_df['Annual Income'] - 12 * loans_df['Monthly Debt']
loans_df.head()
```

```python
# Get the Actual Annual Income of the loan with the ID = 76fa89b9-e6a8-49af-afa1-8151315aba8e
loans_df_loan_id2 = loans_df.loc[loans_df['Loan ID'] == '76fa89b9-e6a8-49af-afa1-8151315aba8e', 'Actual Annual Income'] # with .loc we can squeeze this into one liner, gets the rows where loan id matches to the wanted one and then gets the actual annual income. loan id is not unique apparently
loans_df_loan_id2
```

```python
# Get the Loan ID of the loan with the smallest Actual Annual Income
loans_df_aai_loan_id = loans_df.loc[loans_df['Actual Annual Income'].min(), 'Loan ID']
loans_df_aai_loan_id

```

```python
# How many loans are "Long term"?, so count of of the Term == 'Long Term'
#loans_df_long_term_loans_count = loans_df[loans_df['Term'] == 'Long Term'].count() # with this we got a return of all the rows 
loans_df_long_term_loans_count = len(loans_df[loans_df['Term'] == 'Long Term']) # len() just gives us the rowcount
loans_df_long_term_loans_count
```

```python
# How many loaners have more than 1 bankruptcy? Bankruptcies > 1
loans_df_bankruptcies_count = len(loans_df[loans_df['Bankruptcies'] > 1])
loans_df_bankruptcies_count
```

```python
# How many Short Term loans are for Home Improvements? Purpose + Term columns
loans_df_short_term_loans = len( loans_df[ (loans_df['Purpose'] == 'Home Improvements') & (loans_df['Term'] == 'Short Term') ] )
loans_df_short_term_loans
```

```python
#How many unique loan purposes are there? !!!unique() method returns array of all vals and nunique() method returns count
loans_df_unique_purposes = loans_df['Purpose'].nunique()
loans_df_unique_purposes
```

```python
# What are the 3 most common loan purposes?
loans_df_most_common_purposes = loans_df['Purpose'].value_counts().head(3) # with .head() we can limit the result to top 3
loans_df_most_common_purposes
```

```python
# Is there a correlation between Annual Income and Number of Open Accounts 
# or is there a correlation between Number of Credit Problems and Bankruptcies?
#Explain your findings in your Jupyter notebook also.

# CORRELATION 1 -> PERFECT CORR, 0 -> NO CORRELATION, -1 -> PERFECT NEGATIVE CORRELATION

corr_between_ai_and_noa = loans_df['Annual Income'].corr(loans_df['Number of Open Accounts'])
corr_between_ai_and_noa
# RESULT: 0.143, there is close to none correlation which suggests that these statistic don't have a relation between eachother

corr_between_ncp_and_bankruptcies = loans_df['Number of Credit Problems'].corr(loans_df['Bankruptcies'])
corr_between_ncp_and_bankruptcies

```
