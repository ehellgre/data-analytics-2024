```python
# READ THE CSV FILE + REMOVE CUSTOMER ID COLUMN

loans_data = pd.read_csv('loans.csv')
#loans_data

loans_df = pd.DataFrame(loans_data) 
#loans_df

loans_df = loans_df.drop('Customer ID', axis=1) #drop the customerid from columns
loans_df
```

```python
# PRINT THE HEAD OF THE DATA
loans_df.head()
```

```python
# REMOVE ROWS FROM DATA THAT HAVE A TOO LARGE OF A LOAN

loans_df = loans_df[loans_df['Current Loan Amount'] < 99999999] # bool check if the Current Loan Amount is smaller than 99999999 and if not then removes the row
loans_df
```

```python
# Remove rows that have the annual income as NaN (not a number)
loans_df = loans_df.dropna(subset=['Annual Income']) # subset specifies that this applies only to Annual Income column
loans_df.head()
```

```python
# Get the average Current Loan Amount
loans_df_mean = loans_df['Current Loan Amount'].mean() # self explanatory :D
loans_df_mean
```

```python
# Get the highest and lowest Annual Income in the dataset
loans_df_highest_AI = loans_df['Annual Income'].max()
loans_df_lowest_AI = loans_df['Annual Income'].min()

print(f"highest: {loans_df_highest_AI}, min: {loans_df_lowest_AI}")
```

```python
# Get the Home Ownership value of the Loan ID = bbf87a87-22cd-4d10-bd9b-7a9cc1b6e59d
loans_df_loan_id = loans_df[loans_df['Loan ID'] == 'bbf87a87-22cd-4d10-bd9b-7a9cc1b6e59d'] # first take the wanted loan id row to new variable

loans_df_home_ownership_by_loan_id = loans_df_loan_id['Home Ownership'].iloc[0] #iloc[0] selects the first index assuming there is only 1 when loan id == unique LATER FOUND OUT LOAN ID IS NOT UNIQUE, not gonna change this took too much time to find the way :D
loans_df_home_ownership_by_loan_id
```

```python
# Create a new field into your dataset called Actual Annual Income. (formula: Annual Income – 12 * Monthly Debt)
loans_df['Actual Annual Income'] = loans_df['Annual Income'] - 12 * loans_df['Monthly Debt']
loans_df.head()
```

```python
# Get the Actual Annual Income of the loan with the ID = 76fa89b9-e6a8-49af-afa1-8151315aba8e
loans_df_loan_id2 = loans_df.loc[loans_df['Loan ID'] == '76fa89b9-e6a8-49af-afa1-8151315aba8e', 'Actual Annual Income'] # with .loc we can squeeze this into one liner, gets the rows where loan id matches to the wanted one and then gets the actual annual income. loan id is not unique apparently
loans_df_loan_id2
```

```python
# Get the Loan ID of the loan with the smallest Actual Annual Income
loans_df_aai_loan_id = loans_df.loc[loans_df['Actual Annual Income'].min(), 'Loan ID']
loans_df_aai_loan_id

```

```python
# How many loans are "Long term"?, so count of of the Term == 'Long Term'
#loans_df_long_term_loans_count = loans_df[loans_df['Term'] == 'Long Term'].count() # with this we got a return of all the rows 
loans_df_long_term_loans_count = len(loans_df[loans_df['Term'] == 'Long Term']) # len() just gives us the rowcount
loans_df_long_term_loans_count
```

```python
# How many loaners have more than 1 bankruptcy? Bankruptcies > 1
loans_df_bankruptcies_count = len(loans_df[loans_df['Bankruptcies'] > 1])
loans_df_bankruptcies_count
```

```python
# How many Short Term loans are for Home Improvements? Purpose + Term columns
loans_df_short_term_loans = len( loans_df[ (loans_df['Purpose'] == 'Home Improvements') & (loans_df['Term'] == 'Short Term') ] )
loans_df_short_term_loans
```

```python
#How many unique loan purposes are there? !!!unique() method returns array of all vals and nunique() method returns count
loans_df_unique_purposes = loans_df['Purpose'].nunique()
loans_df_unique_purposes
```

```python
# What are the 3 most common loan purposes?
loans_df_most_common_purposes = loans_df['Purpose'].value_counts().head(3) # with .head() we can limit the result to top 3
loans_df_most_common_purposes
```

```python
# Is there a correlation between Annual Income and Number of Open Accounts 
# or is there a correlation between Number of Credit Problems and Bankruptcies?
#Explain your findings in your Jupyter notebook also.

# CORRELATION 1 -> PERFECT CORR, 0 -> NO CORRELATION, -1 -> PERFECT NEGATIVE CORRELATION

corr_between_ai_and_noa = loans_df['Annual Income'].corr(loans_df['Number of Open Accounts'])
corr_between_ai_and_noa
# RESULT: 0.143, there is close to none correlation which suggests that these statistic don't have a relation between eachother

corr_between_ncp_and_bankruptcies = loans_df['Number of Credit Problems'].corr(loans_df['Bankruptcies'])
corr_between_ncp_and_bankruptcies
# RESULT: 0.747, there is good correlation between Number of Credit Problems and Bankruptcies, which implies that number of credit problems increases the chance of bankruptcies and vice versa

```

```python
salaries_data = pd.read_csv('data_salaries_india.csv')
salaries_df = pd.DataFrame(salaries_data)
salaries_df
```

```python
# Before we can do anything with the salaries, we have to convert them into something more usable. 

# !!!!! Exercise says the salaries can be yearly, monthly or hourly salaries, but quick ctrl+f of the csv file reveals that there is only yr so the moodle function is pointless for this?
# Can just .replace() the indian rupee sign and the commas and the salaries column data is fine?

salaries_df['Salary Per Year'] = salaries_df['Salary'].str.replace('₹', '').str.replace(',', '').str.replace('/yr', '') # i created a new column for stating that the salaries are per year and removed the /yr out of the row data

salaries_df_print_salary_per_year = salaries_df['Salary Per Year'].head(2)
salaries_df_print_salary_per_year

# also lets remove the old Salary column
salaries_df = salaries_df.drop('Salary', axis=1)
salaries_df.head(3)
```

```python
# What are the most common values in different fields (Job Titles, Companies, Location)? Based on the distribution, is the data balanced or not?

#
most_common_job_titles = salaries_df['Job Title'].value_counts().head(20)
most_common_companies = salaries_df['Company Name'].value_counts().head(20)
most_common_locations = salaries_df['Location'].value_counts().head(20)

# check the distribution on job titles
most_common_job_titles
#RESULT: Even though the data heavily upsides the Data Scientist/Data Analyst/Data Engineer/ML Engineer, i'd say the data is still balanced, because
#        pretty much every job title implies the same thing.

# check the distribution on companies
most_common_companies
# RESULT: Data is somewhat balanced. Even though the Tata Consultancy Services dominates the list, the drop off to the botton is smooth.

# check the distribution on locations
most_common_locations
# RESULT: There are only 5 locations, Bangalore has double versus everyone else, but it is still pretty balanced.
```

```python
# Are there any outliers in the data that might affect the averages negatively (certain salaries)? Manage the outliers as you best see fit (either remove them or leave them, based on your analysis)

# convert the Salary Per Year rows to numeric values
salaries_df['Salary Per Year'] = pd.to_numeric(salaries_df['Salary Per Year'])

# calculate the outliers using the z-score ( z = (x - mean) / std )
z = np.abs(stats.zscore(salaries_df['Salary Per Year']))

# identify outliers with zscore greater than 3
threshold = 3
salary_outliers = salaries_df[z > threshold]

# check the outliers (for testing lets take 1, Salary Per Year: 7481517)
salary_outliers
# check the count for the outliers
salary_outliers_len = len(salary_outliers['Salary Per Year'])
salary_outliers_len

# create a new df without the outliers
salaries_df_removed_outliers = salaries_df[z <= threshold]
salaries_df_removed_outliers
```

```python
# Factorize the Role-column, and add the new column to the DataFrame.

#label1, unique1 = pd.factorize(['b', 'd', 'd', 'c', 'a', 'c', 'a', 'b'], sort = True)
#1 print("\n\nNumeric Representation : \n", label1) 
#2 print("Unique Values : \n", unique1) 
# 1 => [1 3 3 2 0 2 0 1]
# 2 => ['a' 'b' 'c' 'd'] (sort = False => [b d c a])

label_role, unique_role = pd.factorize(salaries_df_removed_outliers['Role'])
salaries_df_removed_outliers['Factorized Role'] = label_role
salaries_df_removed_outliers
```

```python
#Finally, check out the correlations. Does anything correlate with anything? Can we make any assumptions?
# df.corr(method="spearman")

# take only numeric values
salaries_df_removed_outliers_numeric_vals = salaries_df_removed_outliers.select_dtypes(include='number')
salaries_df_removed_outliers_numeric_vals

check_correlations = salaries_df_removed_outliers_numeric_vals.corr(method="spearman")
check_correlations

# RESULT: Factorized Role and Salary Per Year has weak corr, but there is some
#         Everything else has a negative correlation, which means that there is not correlation at all (excluding the matching pairs like Salaries Reported - Salaries Reported)
```
